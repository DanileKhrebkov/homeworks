```python
#MODEL
class Shoe:
    GENDER_CHOICES = ('мужская', 'женская')
    TYPE_CHOICES = ('кроссовки', 'сапоги', 'сандалии', 'туфли', 'ботинки', 'шлепанцы')
    
    def __init__(self, shoe_type, gender, color, price, manufacturer, size):
        self.shoe_type = shoe_type
        self.gender = gender
        self.color = color
        self.price = price
        self.manufacturer = manufacturer
        self.size = size
    
    @property
    def shoe_type(self):
        return self._shoe_type
    
    @shoe_type.setter
    def shoe_type(self, value):
        if value not in self.TYPE_CHOICES:
            raise ValueError(f"Недопустимый тип обуви. Допустимые значения: {', '.join(self.TYPE_CHOICES)}")
        self._shoe_type = value
    
    @property
    def gender(self):
        return self._gender
    
    @gender.setter
    def gender(self, value):
        if value not in self.GENDER_CHOICES:
            raise ValueError(f"Недопустимый пол. Допустимые значения: {', '.join(self.GENDER_CHOICES)}")
        self._gender = value
    
    @property
    def price(self):
        return self._price
    
    @price.setter
    def price(self, value):
        if not isinstance(value, (int, float)) or value <= 0:
            raise ValueError("Цена должна быть положительным числом")
        self._price = value
    
    @property
    def size(self):
        return self._size
    
    @size.setter
    def size(self, value):
        if not isinstance(value, (int, float)) or value <= 0:
            raise ValueError("Размер должен быть положительным числом")
        self._size = value
    
    def __str__(self):
        return (f"Обувь: {self.shoe_type}, {self.gender}, цвет: {self.color}, "
                f"размер: {self.size}, цена: {self.price}, производитель: {self.manufacturer}")
```

```python
#View
class ShoeView:
    @staticmethod
    def display_shoe(shoe):
        print(shoe)
    
    @staticmethod
    def display_all_shoes(shoes):
        if not shoes:
            print("Список обуви пуст")
            return
        for i, shoe in enumerate(shoes, 1):
            print(f"{i}. {shoe}")
    
    @staticmethod
    def display_message(message):
        print(message)
    
    @staticmethod
    def get_shoe_input():
        print("\nДобавить новую обувь:")
        shoe_type = input("Тип обуви (кроссовки, сапоги, сандалии, туфли и т.д.): ")
        gender = input("Пол (мужская/женская): ")
        color = input("Цвет: ")
        price = input("Цена: ")
        manufacturer = input("Производитель: ")
        size = input("Размер: ")
        
        return {
            'shoe_type': shoe_type,
            'gender': gender,
            'color': color,
            'price': price,
            'manufacturer': manufacturer,
            'size': size
        }
    
    @staticmethod
    def get_menu_choice():
        print("\nМеню:")
        print("1. Добавить обувь")
        print("2. Показать всю обувь")
        print("3. Показать мужскую обувь")
        print("4. Показать женскую обувь")
        print("5. Показать обувь по типу")
        print("6. Показать обувь по производителю")
        print("7. Удалить обувь")
        print("0. Выход")
        return input("Выберите действие: ")
    
    @staticmethod
    def get_filter_input(prompt):
        return input(prompt)
    
    @staticmethod
    def get_remove_index():
        return int(input("Введите номер обуви для удаления: ")) - 1
```

```python
#MAIN MVC
#from ShoeController import ShoeController
#from ShoeView import ShoeView

def main():
    controller = ShoeController()
    view = ShoeView()
    
    while True:
        choice = view.get_menu_choice()
        
        if choice == '1':
            shoe_data = view.get_shoe_input()
            try:
                price = float(shoe_data['price'])
                size = float(shoe_data['size'])
                success, message = controller.add_shoe(
                    shoe_data['shoe_type'],
                    shoe_data['gender'],
                    shoe_data['color'],
                    price,
                    shoe_data['manufacturer'],
                    size
                )
                view.display_message(message)
            except ValueError:
                view.display_message("Ошибка: цена и размер должны быть числами")
        
        elif choice == '2':
            shoes = controller.get_all_shoes()
            view.display_all_shoes(shoes)
        
        elif choice == '3':
            shoes = controller.get_shoes_by_gender('мужская')
            view.display_all_shoes(shoes)
        
        elif choice == '4':
            shoes = controller.get_shoes_by_gender('женская')
            view.display_all_shoes(shoes)
        
        elif choice == '5':
            shoe_type = view.get_filter_input("Введите тип обуви для поиска: ")
            shoes = controller.get_shoes_by_type(shoe_type)
            view.display_all_shoes(shoes)
        
        elif choice == '6':
            manufacturer = view.get_filter_input("Введите производителя для поиска: ")
            shoes = controller.get_shoes_by_manufacturer(manufacturer)
            view.display_all_shoes(shoes)
        
        elif choice == '7':
            shoes = controller.get_all_shoes()
            view.display_all_shoes(shoes)
            if shoes:
                try:
                    index = view.get_remove_index()
                    removed_shoe = controller.remove_shoe(index)
                    if removed_shoe:
                        view.display_message(f"Удалена обувь: {removed_shoe}")
                    else:
                        view.display_message("Неверный номер обуви")
                except ValueError:
                    view.display_message("Ошибка: введите число")
        
        elif choice == '0':
            print("Выход из программы")
            break
        
        else:
            view.display_message("Неверный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()

def main():
    controller = ShoeController()
    view = ShoeView()
    
    while True:
        choice = view.get_menu_choice()
        
        if choice == '1':
            shoe_data = view.get_shoe_input()
            try:
                price = float(shoe_data['price'])
                size = float(shoe_data['size'])
                success, message = controller.add_shoe(
                    shoe_data['shoe_type'],
                    shoe_data['gender'],
                    shoe_data['color'],
                    price,
                    shoe_data['manufacturer'],
                    size
                )
                view.display_message(message)
            except ValueError:
                view.display_message("Ошибка: цена и размер должны быть числами")
        
        elif choice == '2':
            shoes = controller.get_all_shoes()
            view.display_all_shoes(shoes)
        
        elif choice == '3':
            shoes = controller.get_shoes_by_gender('мужская')
            view.display_all_shoes(shoes)
        
        elif choice == '4':
            shoes = controller.get_shoes_by_gender('женская')
            view.display_all_shoes(shoes)
        
        elif choice == '5':
            shoe_type = view.get_filter_input("Введите тип обуви для поиска: ")
            shoes = controller.get_shoes_by_type(shoe_type)
            view.display_all_shoes(shoes)
        
        elif choice == '6':
            manufacturer = view.get_filter_input("Введите производителя для поиска: ")
            shoes = controller.get_shoes_by_manufacturer(manufacturer)
            view.display_all_shoes(shoes)
        
        elif choice == '7':
            shoes = controller.get_all_shoes()
            view.display_all_shoes(shoes)
            if shoes:
                try:
                    index = view.get_remove_index()
                    removed_shoe = controller.remove_shoe(index)
                    if removed_shoe:
                        view.display_message(f"Удалена обувь: {removed_shoe}")
                    else:
                        view.display_message("Неверный номер обуви")
                except ValueError:
                    view.display_message("Ошибка: введите число")
        
        elif choice == '0':
            print("Выход из программы")
            break
        
        else:
            view.display_message("Неверный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
